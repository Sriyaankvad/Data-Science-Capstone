```python
import pandas as pd
from pathlib import Path
```

# Data Loading

Read csv file

```python
df = pd.read_csv("airbnb_v2.csv", header = 1, skipinitialspace = True)
df
```

List columns (for convenience)

```python
def list_columns():
    for col in df:
        print(col)
        
list_columns()
```

# Data Cleaning

Remove Unnamed Column 15

```python
df.drop(df.columns[15], axis = 1, inplace=True)
df
```

Remove quotes in "name" column

```python
df.rename(columns = {'\"name\"' : 'name'}, inplace = True)
list_columns()
```

Remove all Missing Values

```python
df.dropna(axis=0, inplace = True)
df
```

Change Month and Week in Minimum Nights to Nights

```python
# Find out all unique types of time periods (nights, weeks, months etc)
def add_text(x):
    if ' ' not in x:
        x += " nights"
    return x

df['minimum nights'] = df['minimum nights'].apply(add_text)
df['time period'] = df['minimum nights'].str.split(' ').str[1]
df['minimum nights'] = df['minimum nights'].str.split(' ').str[0]

df
```

```python
sorted(df['time period'].unique())
```

```python
# Only two time periods are nights (1 night) and month (31 nights). Apply this to minimum nights
df['minimum nights'] = pd.to_numeric(df['minimum nights'])
df['time period'].replace({'nights' : 1, 'month' : 31}, inplace = True)
df['minimum nights'] = df['minimum nights'] * df['time period']
del df['time period']
df
```

Remove all non-numeric keys from numeric columns

```python
numeric_column_names = ['ID', 'host [id]', 'latitude', 'longitude', 'number_of_reviews', 
                        'reviews_per_month','noise(dB)', 'Price ($)']

for col in numeric_column_names:
    # Longitude is the only column with negative values, and all values are negative, so will ignore for now
    df[col].replace('[^0-9\.]', '', inplace = True, regex = True)
    df[col] = pd.to_numeric(df[col])

# Add negative values back to longitude
df['longitude'] = df['longitude'].apply(lambda x: x*-1)
df
```

Changing Floor to All Numeric

```python
sorted(df['floor'].unique())
```

```python
floor_to_num = {'1st-floor' : '1', '5th Floor' : '5', 'Eighth Floor' : '8', 'Eleventh Floor' : '11', 
                'Fifth Floor' : '5', 'First Floor' : '1', 'First Floor (Balcony View)' : '1', 
                'Fourth Floor' : '4', 'Hundreth Floor' : '100', 'Lower Level' : '0', 'Ninth Floor' : '9', 
                'Second Floor' : '2', 'Seventeenth Floor' : '17', 'Seventh Floor' : '7', 'Sixteenth Floor' : '16', 
                'Sixth Floor' : '6', 'Tenth Floor' : '10', 'Third Floor' : '3', 'Thirteenth Floor' : '13', 
                'Twentieth Floor' : '20'}

df['floor'].replace(floor_to_num, inplace=True)
df['floor'] = pd.to_numeric(df['floor'])

sorted(df['floor'].unique())
```

```python
df
```

Fix Neighborhood Group

```python
sorted(df['neighbourhood group'].unique())
```

```python
df['neighbourhood group'] = df['neighbourhood group'].str.split(' ').str[0]
df['neighbourhood group'] = df['neighbourhood group'].str.split('-').str[0]
sorted(df['neighbourhood group'].unique())
```

```python
neighbourhood_group_fixing = {'**Manhattan**' : 'Manhattan', 'Staten' : 'Staten Island', 'brklyn' : 'Brooklyn', 
                             'manhattan' : 'Manhattan', 'mnahattan' : 'Manhattan', 'quens' : 'Queens'}
df['neighbourhood group'].replace(neighbourhood_group_fixing, inplace=True)
sorted(df['neighbourhood group'].unique())
```

```python
df
```

Fixing Neighbourhood

```python
sorted(df['neighbourhood'].unique())
```

```python
# Most of the values with '-' in it seem to have neighbourhood group following the '-', which is irrelavent
# Testing this, all but Co-op City follow this rule
sorted(df[df['neighbourhood'].str.contains('-')]['neighbourhood'].unique())
```

```python
# Fix this issue
df['neighbourhood'] = df['neighbourhood'].str.split('-').str[0]
df['neighbourhood'].replace('Co', 'Co-op City', inplace = True)
sorted(df['neighbourhood'].unique())
```

```python
# 'Bay Terrace, Staten Island' still has irrelavent information, so fix this
df['neighbourhood'].replace('Bay Terrace, Staten Island', 'Bay Terrace', inplace = True)
sorted(df['neighbourhood'].unique())
```

```python
df
```

Fixing room and type

```python
sorted(df['room and type'].unique())
```

```python
fix_rooms = {'Complete Home/Apt': 'Entire Home/Apartment', 'Entire Home/Apt' : 'Entire Home/Apartment',
            'Entire home/apt' : 'Entire Home/Apartment', 'Private room' : 'Private Room', 
            'Room Type - Entire Home/Apartment' : 'Entire Home/Apartment', 
            'Room Type Private' : 'Private Room', 'Room Type-Shared room' : 'Shared room'}
df['room and type'].replace(fix_rooms, inplace=True)
sorted(df['room and type'].unique())
```

```python
df
```

Fix Name and Host-name columns

```python
# Remove special characters
cols = {'name', 'host-name'}

for col in cols:
    df[col].replace('[^0-9A-Za-z &/]', '', regex = True, inplace = True)
    
df
```

```python
# Remove lines with wrong characters
def replace_special_chars(txt):
    txt = txt.strip()
    if txt.isnumeric() or txt == '/' or txt == '&':
        return ''
    return txt

for col in cols:
    df[col] = df[col].apply(replace_special_chars)
    df.drop(df.loc[df[col]==''].index, inplace=True)

df
```
