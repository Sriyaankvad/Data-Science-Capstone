# Cleaning Data

### Remove NaN rows from locs

```python
locs = locs.dropna()
locs = locs.reset_index(drop = True)
locs
```

### Add "Distance from POI" column to df

This column has a value $distance = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$ where $x_1$ = latitude of location, $x_2$ = latitude of closest point of interest, $y_1$ = longitude of location, $y_2$ = longitude of closest point of interest

Turning the dataset into an array containing the latitude and longitudes in numerical form

```python
# Turning the dataset into an array containing the latitude and longitudes in numerical form
locs_array = locs['Coordinates'].to_numpy()
def apply(a):
    for i in range(0, len(a)):
        x = a[i]
        lat = float(x[0: x.index('Â°')])
        long = -1*float(x[x.index(',')+2:x.rindex('Â°')])
        a[i] = [lat, long]

apply(locs_array)
locs_array
```

Calculate Distances

```python
def get_distance(lat, long):
    distance = float("inf")
    for coords in locs_array:
        distance = min(((lat - coords[0])**2 + (long-coords[1])**2)**0.5, distance)
    return distance

df['distance_from_poi'] = df.apply(lambda row: get_distance(float(row.latitude), float(row.longitude)), axis=1)
df
```

### Add crime rates to df

Essentially merging the NYC_crime_statistics dataset with the main dataset

Extract zipcodes from address. Drop any column that doesn't provide a zipcode

```python
df['zipcode'] = df.apply(lambda row: row.Location[-20:-15], axis=1)
df['zipcode'] = np.where(df['zipcode'].str.isnumeric(), df['zipcode'], 0)
df.drop(df.loc[df['zipcode']==0].index, inplace=True)
df['zipcode'] = df.apply(lambda row: int(row.zipcode), axis=1)
df
```

Adding the crime rates for all possible zipcodes. Has NaN value if the zipcode doesn't have a recorded crime rate

```python
crime_dict = dict(zip(crime['Zip Codes'], crime['TOTAL SEVEN MAJOR FELONY OFFENSES']))
df['crime_rates'] = df.apply(lambda row: crime_dict[row.zipcode] if row.zipcode in crime_dict else np.nan, axis=1)
df
```

# Hypothesis Testing

### Get sample of dataframe

```python
df_sample = df.sample(frac = 0.1, random_state = 420)
df_sample
```

### Although we said that there isn't a correlation between crime rate and price, there may be relation between price and whether there is crime or not, so let's test that.

Let $\mu_{0}$ = Mean log price for rooms in places with crime

Let $\mu_{1}$ = Mean log price for rooms in places without crime

$H_{o}$: $\mu_{0}$ - $\mu_{1}$ = 0

$H_{a}$: $\mu_{0}$ - $\mu_{1}$ â‰  0

```python
has_crime = df_sample.loc[df['has_crime'] == "Yes"]['log_price'].values
crimeless = df_sample.loc[df['has_crime'] == "No"]['log_price'].values

has_crime = has_crime[~np.isnan(has_crime)]
crimeless = crimeless[~np.isnan(crimeless)]

len(has_crime), len(crimeless)
```

Both sets have more than 30 values, so we can perform theoretical test

```python
ztest, pval1 = stests.ztest(has_crime, crimeless, value=0,alternative='two-sided')
ztest, float(pval1)
```

At the 5% level, p < alpha so we reject the null hypothesis. The data shows significant evidence that the Mean log price for rooms in places with crime differs from the Mean log price for rooms in places without crime. Hence, we can say that having crime plays an impact on the log price

### In the box plots for log price vs room type, it appeared as though the mean log price depended on the room type. Let's test this

Let $\mu_{0}$ = Mean log price for Entire home/apt room type

Let $\mu_{1}$ = Mean log price for Private room room type

$H_{o}$: $\mu_{0}$ - $\mu_{1}$ = 0

$H_{a}$: $\mu_{0}$ - $\mu_{1}$ â‰  0

```python
entire = df_sample.loc[df_sample['room_type'] == 'Entire home/apt']['log_price'].values
private = df_sample.loc[df_sample['room_type'] == 'Private room']['log_price'].values

entire = entire[~np.isnan(entire)]
private = private[~np.isnan(private)]

len(entire), len(private)
```

Both sets have more than 30 values, so we can perform theoretical test

```python
ztest, pval1 = stests.ztest(entire, private, value=0,alternative='two-sided')
ztest, float(pval1)
```

At the 5% level, p < alpha so we reject the null hypothesis. The data shows significant evidence that the Mean log price for Entire home/apt room type differs from the Mean log price for Private room room type. Hence, we can say that the room price plays an impact on the log price
